```python
bali.dropna(axis=0,how="all",inplace=True)
bali
```

```python
bali.drop(bali.index[0],inplace=True)
bali.set_index('Unnamed: 0',inplace=True)
```

```python
x,y=0,36
for i in range(5):
    bali.drop(bali.iloc[:,x:y], inplace=True, axis=1)
    x+=16
    y=x+36
```

```python
bali1=bali.copy()
bali1.sort_index(axis=1, inplace=True)
bali=bali1.T
```

```python
sorted_bali=pd.DataFrame()
sorted_bali['values']=[]
sorted_bali['segment']=[]
sorted_bali['period']=[]
sorted_bali['subtype']=[]
sorted_bali['date']=[]
```

```python
x=pd.DataFrame()
for i in bali.index:
    x['values']= bali.loc[i][1:]
    x['segment']=i.split('-')[0].strip()
    x['period']=i.split('-')[1].strip()
    x['subtype']=bali.loc[i][0]
    x['date']=bali.columns[1:]
    sorted_bali=sorted_bali.append(x)
```

```python
sorted_bali.reset_index(inplace=True)
sorted_bali.drop(columns=['index'],inplace=True)
```

```python
def find_outliers_IQR(b):
                       q1=b.quantile(0.25)
                       q3=b.quantile(0.75)
                       IQR=q3-q1
                       outliers = b[((b<(q1-1.5*IQR)) | (b>(q3+1.5*IQR)))]
                       return outliers
```
